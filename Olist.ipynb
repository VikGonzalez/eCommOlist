```python
#import dependencies
import pandas as pd
```

```python
# create path to file "olist_order_items_dataset"
csv_path = ("Resources/olist_order_items_dataset.csv")
```

```python
# create variable to file
order_items = pd.read_csv(csv_path)
```

```python
# print head of file
order_items.head()
```

```python
# print shape to confirm correct extraction of original file
order_items.shape
```

```python
# create path to file "olist__products_dataset"
csv_path_2 = ("Resources/olist_products_dataset.csv")
```

```python
# create variable to file
product_dataset = pd.read_csv(csv_path_2)
```

```python
# print head of file
product_dataset.head()
```

```python
# print shape to confirm correct extraction of original file
product_dataset.shape
```

```python
# merge files based on column "product_id"
merge_df = pd.merge(order_items,product_dataset, on = "product_id",how="inner")
```

```python
# print head
merge_df.head()
```

```python
# print shape to confirm correct merge
merge_df.shape
```

```python
# print shape to confirm correct length
len(merge_df["product_category_name"])
```

```python
# extract unique values based on column: "product_category_name"
merge_df["product_category_name"].unique()
```

```python
# insert new column for "total_order_values" (multiplication of order_item_id and price)
merge_df["Total_order_values"]=""
```

```python
merge_df.head()
```

```python
# insert function lambda for multiplication and insert results into column new column: "total_order_values"
merge_df["Total_order_values"] =merge_df.apply(lambda row: row[1]*row[5], axis=1)
```

```python
# print head
merge_df.head()
```

```python
# insert new column for "total_order_values" (multiplication of order_item_id and freight_value)
merge_df["Total_freight_values"]=""
```

```python
# insert function lambda for multiplication and insert results into column new column: "total_freight_values"
merge_df["Total_freight_values"] =merge_df.apply(lambda row: row[1]*row[6], axis=1)
```

```python
# print head
merge_df. head()
```

```python
# insert new column for "total_revenue_by_order" (sum of total_order_values plus total_freight_values and multiply for 21% of comission)
merge_df["Total_Revenue_by_order"]=""
```

```python
# insert function lambda for sum and insert results into column new column: "total_revenue_by_order"
merge_df["Total_Revenue_by_order"] =merge_df.apply(lambda row: (row[15]+row[16])*.21, axis=1)
```

```python
# print head
merge_df.head()
```

```python
# drop all columns we won´t need for analysis
df_clean= merge_df.drop(["shipping_limit_date","product_name_lenght","product_description_lenght","product_photos_qty","product_weight_g","product_length_cm","product_height_cm","product_width_cm"],axis=1)
```

```python
# print head
df_clean.head()
```

```python
# create path to file "olist__sellers_dataset"
csv_path_3 = ("Resources/olist_sellers_dataset.csv")

# create variable to file
sellers = pd.read_csv(csv_path_3)
sellers.head()
```

```python
# Merge df_clean.head with sellers file 
df_clean= df_clean.merge(sellers, how= "inner")
```
